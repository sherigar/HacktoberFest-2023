# Stock News Headlines Prediction

```python
import pandas as pd # importing the pandas library
```

### Reading the dataset

```python
# reading the csv data using read_csv function
df=pd.read_csv(r'Data.csv', encoding = "ISO-8859-1")
```

```python
df.head() # label is 1 means stock price will increase w.r.t. particular headline, else if 0 , there either constant are slightly increase
```

### Separating training and testing dataset

```python
# separating the training and testing set of data
train = df[df['Date'] < '20150101'] # 1 jan, 2015
test = df[df['Date'] > '20141231'] # 31 dec, 2014
```

```python
# now we don't want the first two column, i.e. Data and Label column, therefore we insert all the data from column 2 to 27
data=train.iloc[:,2:27] # here : denotes data contains all the rows that are in train set and 2:27, denotes data contains all the data from column 2 to 27 from train set
data
```

```python
data.columns
```

### Cleaning the data

```python
# Here we clean the data by Removing punctuations (removing all blanks apart from a-z)
data.replace("[^a-zA-Z]"," ", inplace=True)
```

```python
data.head()
```

```python
# Renaming column names for ease of access, column 1 renamed to 0, then 1,2,......
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
new_Index
```

```python
data.columns= new_Index
data.head(5)
```

```python
data.index # getting all the index
```

```python
len(data)
```

```python
### accessing data of 1st column
data['0']
```

```python
# Convertng headlines to lower case for easy accessing
#why str.lower on 0,1,2---- bcz we can apply function on a column only
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

### Performing data wrangling and data merging

```python
# now we need to perform data wrangling and data merging and make data ready for modelling purpose
##print entire data of 2nd row
data.iloc[1,0:25]
```

```python
#create a list of all headlines for row 2,as 25 headlines in row 2

# headlines = []
# for i in data.iloc[1,0:25]:
#     headlines.append(i)
#     .join(headlines)

' '.join([str(x) for x in data.iloc[1,0:25]]) # the above code is performed using list comprehension
```

```python
## create a list headlines to insert all the headlines of all rows
# append all the data in one list only headlines
headlines=[]
for row in range(0,len(data)):
    headlines.append(' '.join([str(i) for i in data.iloc[row,0:25]]))
```

```python
headlines[0:3]
```

```python
# now we have to simply pass this data to NLP model, we cannot pass it to ML model, because it is in text format
# therefore by first passing it to the NLP model we convert it to some string format and then pass it to the ML model
```

```python
# Implement Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier # imported RandomForestClassifier
```

```python
cv=CountVectorizer(ngram_range=(2,2))
traindata_x=cv.fit_transform(headlines) # passing traindata_x to randomclassifier
```

```python
# implement RandomForest Classifier, this process will take some time, approximately 2 to 3 min
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindata_x,train['Label'])
```

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
```

```python
test_data = cv.transform(test_transform)
```

```python
randomclassifier.fit(traindata_x,train['Label'])
```

```python
predictions = randomclassifier.predict(test_data)
```

### Importing Library for checking accuracy of the model
